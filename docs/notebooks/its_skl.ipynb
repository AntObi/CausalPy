# Interrupted time series with sci-kit learn models

```python
import pandas as pd
import pathlib
import arviz as az
```

```python
az.style.use("arviz-darkgrid")
```

## Load data

```python
its_data_path = pathlib.Path.cwd().parents[1] / "causalpy" / "data" / "its.csv"
df = pd.read_csv(its_data_path, parse_dates=["date"])
df.set_index("date", inplace=True)
treatment_time = pd.to_datetime("2017-01-01")
```

## Run the analysis

```python
from causalpy.skl_experiments import InterruptedTimeSeries
from sklearn.linear_model import LinearRegression

result = InterruptedTimeSeries(
    df,
    treatment_time,
    formula="y ~ 1 + t + C(month)",
    prediction_model=LinearRegression(),
)
```

## Examine the results

```python
fig, ax = result.plot();
```
